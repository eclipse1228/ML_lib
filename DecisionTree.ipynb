# 결측치 확인

```python
data.isna().mean()
# native-country 는 class의 평균값으로 대체한 상태임 (일반적으로 불가능한데, 트리기반 모델에서는 가능함)
# 결측 데이터를 -9나 -99로 채워주자 (선형모델에서는 왜곡을 불러오지만, 트리에서는 아니다.)
```

```python
data['native-country'] = data['native-country'].fillna(-99)
```

```python
# workclass , occupation 결측치
data['workclass'].value_counts()
# private가 많으니까.. 결측치를 private로 하자!.. 70% 라서 좀 그렇지만..!
```

```python
data['workclass'] = data['workclass'].fillna('Private')
```

```python
data['occupation'].value_counts() # 이렇게 비교적 고를 경우.. Unknown 으로 한다.
```

```python
data['occupation'] = data['occupation'].fillna('Unknown')
```

```python
data = pd.get_dummies(data, drop_first = True) # 더미 변수로 변환
```

```python
data
```

# 모델링 및 평가하기

## 분류 결정 트리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class',axis = 1), data['class'], test_size= 0.4 , random_state = 100)
```

```python
# DecisionTreeRegressor , DecisionTreeClassifier  Choose 1
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

# 오버피팅 문제
마지막 노드에 1개의 관측치만 가도록 하면 100% 정확도 분류가 가능함.   
하지만 그러면 좋은 모델이 아님.   
만약 지나치게 깊은 트리모델을 만들면, 시험셋에 대한 예측은 오히려 떨어지는 경향이 있다. 사실 이것은 모든 머신러닝 모델에서 발생하는 문제이다.

```python
# 너무 깊게 만들면, 훈련셋과 시험셋의 정확도가 달라진다.
# 근데 Default 가 가장 깊게 이다.
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train socre:', accuracy_score(y_train, train_pred), 'Test score', accuracy_score(y_test, test_pred)) #  훈련셋, 시험셋 정확도 평가
```

```python
#  0.16 정도면 오버피팅이 발생했다고 판단한다. 시험셋의 정확도를 올리자.
model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train socre:', accuracy_score(y_train, train_pred), 'Test score', accuracy_score(y_test, test_pred)) #  훈련셋, 시험셋 정확도 평가
```

```python
model = DecisionTreeClassifier(max_depth=7)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train socre:', accuracy_score(y_train, train_pred), 'Test score', accuracy_score(y_test, test_pred)) #  훈련셋, 시험셋 정확도 평가
```

# Tree Visualization

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30,15)) # graph size
plot_tree(model)
plt.show() # 불필요한 문자가 나오지 않게 출력
```

```python
# 3단계 까지만 확대해서 보기
plt.figure(figsize=(30,15))
plot_tree(model , max_depth=3, fontsize=15, feature_names= X_train.columns) # feature... : 변수 이름을 추가하여 그래프 출력
plt.show()
```
