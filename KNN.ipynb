# Min-Max Scaling

```python
mm_scaler = MinMaxScaler() # Scaling object
mm_scaled = mm_scaler.fit_transform(data) # execute Scaling
mm_scaled = pd.DataFrame(mm_scaled, columns = data.columns)
round(mm_scaled.describe(),2)
```

# Apply Scaling
1. 종속변수는 스케일링 대상에서 제외해야한다.
2. 스케일링 전, 훈련셋,시험셋 나누기

```python
from sklearn.model_selection import train_test_split
X = data.drop('class', axis =1) # 독립변수
y = data['class']
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size =0.2 , random_state = 100)
```

```python
mm_scaler = MinMaxScaler() # obj
# mm_scaler.fit(X_train) # train
```

```python
# X_train_scaled = mm_scaler.transform(X_train) # 학습셋 트랜스폼
# X_test_scaled = mm_scaler.transform(X_test) # 시험셋 트랜스폼
```

```python
X_train_scaled = mm_scaler.fit_transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

# 모델링 및 예측 및 평가하기 (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier() # model obj
knn.fit(X_train_scaled,y_train) # train
pred = knn.predict(X_test_scaled) # predict
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

# HyperParameter Tuning
n_neighbors 의 매개변수는 예측에 가까운 이웃을 몇개나 고려할지 정한다.
Default 5

```python
knn = KNeighborsClassifier(n_neighbors= 7) # model obj
knn.fit(X_train_scaled,y_train) # train
pred = knn.predict(X_test_scaled) # predict
accuracy_score(y_test, pred)
```

```python
# 반복문으로 찾자
scores = []
for i in range(1, 21):
  knn = KNeighborsClassifier(n_neighbors= i) # model obj
  knn.fit(X_train_scaled,y_train) # train
  pred = knn.predict(X_test_scaled) # predict
  acc = accuracy_score(y_test, pred)
  scores.append(acc)

scores
```

```python
sns.lineplot(x=range(1,21), y=scores)
```

```python
# 13 is best
knn = KNeighborsClassifier(n_neighbors=13) # model obj
knn.fit(X_train_scaled,y_train) # train
pred = knn.predict(X_test_scaled) # predict
accuracy_score(y_test, pred)
```

# KNN 매개변수 Weights
# 동점일 때 거리가 더 가까운쪽으로 결정하도록 할 수 있음



```python
# 반복문으로 찾자
scores = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors= i,weights='distance') # model obj
    knn.fit(X_train_scaled,y_train) # train
    pred = knn.predict(X_test_scaled) # predict
    acc = accuracy_score(y_test, pred)
    scores.append(acc)

scores
```

```python
sns.lineplot(x=range(1,21), y=scores)
```
