```python
left = pd.read_csv('left.csv')
right = pd.read_csv('right.csv')
```

```python
left
```

```python
right
```

```python
# left.merge(right, how= 'outer') # 중복 되지 않는 고유한 값 추가
left.merge(right, how='left') # 없는거는 NaN 값으로 처리됨
```

```python
# join() 함수  // index 인덱스를 기준으로 합친다. 'key'라는 변수가 겹쳐서 오류
# left.join(right)
# left.drop('key', axis =1).join(right.drop('key', axis=1))
```

```python
# key 변수를 key 역활을 하게 index 로 만들어준다.
left = left.set_index('key')
right = right.set_index('key')
left.join(right)
```

```python
left.join(right,how='inner')
```

```python
# concat(리스트)  행 기준으로 결합된다. 얘는 pd. 으로 시작함.
pd.concat([left,right])
# concat 마지막 행 아래에 결합시킨다. 즉 단순하게 df 2개를 끝나는 지점에 붙인거임.
```
